```python
import pandas as pd
import yfinance as yf
from sqlalchemy import create_engine
import ta 
```

```python
engine = create_engine('sqlite:///Section 14 Video 72 SNP 500 Stocks DataBase.db')
```

```python
names = pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'", engine)
names = names['name'].to_list()
```

```python
framelist = []

for name in names:
    framelist.append(pd.read_sql(f'SELECT Date, Close FROM "{name}"', engine).set_index('Date'))
```

```python
framelist 
```

```python
import ta.momentum
import ta.trend

def macd_decision(df):
    df['MACD_diff'] = ta.trend.macd_diff(df['Close'])
    df.loc[ (df['MACD_diff'] > 0) & (df['MACD_diff'].shift(1) < 0),'Decision'] = 'Buy'

def rsi_decision (df):
    df['RSI'] = ta.momentum.rsi(df['Close'], window = 10)
    df['SMA200'] = ta.trend.sma_indicator(df['Close'], window = 200)
    df.loc[(df['Close'] > df['SMA200']) & (df['RSI'] < 30) ,'Decision RSI/SMA'] = 'Buy'
```

```python
for i in framelist:
    macd_decision(i)
    rsi_decision(i)
```
